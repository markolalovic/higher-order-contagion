# Examples of Higher-Order Structures 2

```python
import sys
sys.path.append('../src/')

import networkx as nx
import numpy as np
```

```python
from higher_order_structures import ErdosRenyiSC

from utils_sage import draw_hypergraph
```

```python
# --- Parameters for a small example ---
N_example = 24
d1_target_example = 6.0
d2_target_example = 2.0
```

```python
g_sc = ErdosRenyiSC(N_example, d1_target_example, d2_target_example)
```

```python
g_sc.summary()
```

```python
nx_graph_for_layout = nx.Graph()
nx_graph_for_layout.add_nodes_from(range(g_sc.N))
edges_to_draw = g_sc.get_edges(order=1)
nx_graph_for_layout.add_edges_from(edges_to_draw)
```

```python
positions = nx.spring_layout(nx_graph_for_layout, k=0.8/np.sqrt(g_sc.N))
file_name = "../figures/higher_order_structures/ho_erdos_renyi_sc_spring_layout.svg"
draw_hypergraph(g_sc, positions, file_name)
```

```python
positions = nx.kamada_kawai_layout(nx_graph_for_layout)
file_name = "../figures/higher_order_structures/ho_erdos_renyi_sc_kamada_kawai.svg"
draw_hypergraph(g_sc, positions, file_name)
```
