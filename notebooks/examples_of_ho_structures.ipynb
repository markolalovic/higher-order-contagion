## Example

```python
N = 4
g = HigherOrderStructure(N)
g.name = "Example"
edges = [(0, 1), (1, 2), (2, 0), (2, 3), (3, 1), (1, 2, 3)]
g.set_edges(edges)

# set node positions and draw it:
positions = {0: (0, 1), 1: (0, 0), 2: (1, 1), 3: (1, 0)}
draw_hypergraph(g, pos=positions)
```

```python
g.print()
```

## Cycle Hypergraph

```python
N = 10
g = HigherOrderStructure(N)
g.name = "Cycle Hypergraph"

edges = []
for i in range(N):
    # 2-node edges form a cycle
    j = (i + 1) % N
    edges.append((i, j))
    
    # 3-node edges form a cycle
    k = (i + 2) % N
    edges.append((i, j, k))

g.set_edges(edges)
draw_hypergraph(g)
```

```python
g.print()
```

## (Binomial) Random hypergraph 

```python
N = 10
g = RandomHypergraph(N, 0.5, 0.5 / N)
draw_hypergraph(g)
```

```python
g.print()
```
